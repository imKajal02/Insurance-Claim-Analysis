# Feature Engineering

. Removing outliers from column Annual_Premium

```python
Q1 = df['Annual_Premium'].quantile(0.25)
Q3 = df['Annual_Premium'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
dft = df[~((df['Annual_Premium'] < (Q1 - 1.5 * IQR)) | (df['Annual_Premium'] > (Q3 + 1.5 * IQR)))]
```

```python
dft
```

```python
dft.info()
```
